```python
import urllib.request
from bs4 import BeautifulSoup
import urllib.request 
import pandas as pd
```

```python
req = urllib.request.Request('http://bigdata-trader.com/itemcodehelp.jsp') 
data = urllib.request.urlopen(req).read() 
bs = BeautifulSoup(data, 'html.parser') ##받아온 url parsing 작업
td_Tag = bs.find_all('td') ## 그 중에 td태그만 가져옴

allCoin = []
for i in range(0,1152):
    allCoin.append(td_Tag[10*i+6].find('a')['href'].split('/')[2]) ## td 태그 안에 Coin 이름 정보가 있는 위치를 찾아 allCoin에 저장
```

```python
allCoin
```

```python
allCoinCount = []
print("총 " + str(len(allCoin)) + " 번 돌아갑니다.")
c = 0
for i in allCoin:
    c = c + 1
    if(c % 100 == 0):
        print(str(c) + " 번째 돌아가고 있습니다.")
    url2 = 'https://coinmarketcap.com/currencies/' + i + '/#markets'
    req2 = urllib.request.Request(url2) 
    data2 = urllib.request.urlopen(req2).read() 
    bs2 = BeautifulSoup(data2, 'html.parser')
    l = bs2.find_all('td')
    count = int(len(l)/7) ## 한 마켓에 관해서 td태그가 7번씩 들어가서, 전체 td 갯수에서 7로 나눔으로써 마켓수를 알 수 있습니다
    print(i,count)
    allCoinCount.append([i,count])
```

```python
pd.DataFrame(allCoinCount).to_csv("/Users/hsw/Desktop/개인 파일/스터디/Coin Project/allCoinMarketCount.csv", index = False, header = False, encoding='utf-8')
```
